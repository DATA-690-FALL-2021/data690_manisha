## pandas.melt
- Meltis used to massage the dataframe into a format where one or more columns are identifier variables,while all other columns are considered measured variables.


```python
billboard_melt = pd.melt(
    billboard, 
    id_vars = ['year', 'artist', 'track' ,'time','date.entered'],
    var_name= 'week',
    value_name = 'rating' )
billboard_melt.head()
```

```python
weather = pd.read_csv('data/weather.csv')
weather
```

```python
weather_melt = pd.melt(
    weather,
    id_vars=['id','year','month','element'], 
    var_name='day', 
    value_name= 'temp'
    )                   
weather_melt
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp'
) 
weather_tidy.reset_index()
```

- The data set which we worked as part of our code contains large amount of data where it becomes as tough as for us to read the data and to find the required data.
- We have learned how to work with tab separated value files.
- The below are the various problems with messy datasets along with their remedies.
- Column headers are values, not variable names
- Multiple variables are stored in oone column
- Variables are stored in both rows and columns
- Multiple types of observational units are stored in the same table
- A single observational unit is stored in multiple tables
- Tidy Data helps us to clean the data and organize that data into a readable way.
- This not only cleans the data, it also helps us to decrease the storage that is needed to store the data by perfoming various methods which helps to reduce the time required to read the data that is needed.
- Hence, we were able to visualize all the required data values from the data frame in an organized way.
