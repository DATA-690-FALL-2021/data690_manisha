##Find out the definition of the above 9 variables and document the definitions

### UNITID 
- A unique Id for institution. 

### INSTNM
- It is institution name.

### CITY
- City where the institution is.

### STABBR:
- This refers to the State postcode of the university.

### ZIP
- It is the zip code of the city where the institution is located.

### ADM_RATE
- It is the admission rate of the term across the campus, where it the total of admitted students in the particular term.

### UGDS
- This contains enrollment of undergraduate certificate/degree-seeking students in the university.

### TUITIONFEE_IN
- It defines the In-state tuition and fees of the university.

###MN_EARN_WNE_P10
- It has the Mean earnings of students working and not enrolled for 10 years after entry from the university.

###Find and display UMBC

```python
new_df.set_index("INSTNM", inplace = True)
a = new_df.loc["University of Maryland-Baltimore County"]
a
```

###Find Univ. of Maryland College Park

```python
new_df.set_index("INSTNM",inplace = True)
b = new_df.loc['University of Maryland-College Park']
b
```

###Find and display Loyola College Maryland

```python
new_df.set_index("INSTNM",inplace = True)
c = new_df.loc['Loyola University Maryland']
c
```

###Find and display Johns Hopskins

```python
new_df.set_index("INSTNM",inplace = True)
d = new_df.loc['Johns Hopkins University']
d
```

### Adding all the four colleges as a one dataframe

```python
new_df_final = pd.concat([a, b, c, d], axis=0)
new_df_final
```

###Minimum of in-state tuition fee for above four institutions

```python
min_tuition = new_df_final['TUITIONFEE_IN'].min()
print("Minimum of in state tuition fee:", min_tuition)
```

###Maximum of in-state tuition fee for above four institutions

```python
max_tuition = new_df_final['TUITIONFEE_IN'].max()
print("Maximum of in state tuition fee:", max_tuition)
```

###Average of in-state tuition fee for above four institutions

```python
avg_tuition = new_df_final['TUITIONFEE_IN'].mean()
print("Average of in state tuition fee:", avg_tuition)
```

###Filter the smaller dataset further to only keep colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia.

```python
Five_States_Colleges = new_df[new_df['STABBR'].isin(['MD', 'VA', 'DC', 'DE', 'PA'])]
Five_States_Colleges = pd.DataFrame(Five_States_Colleges)
Five_States_Colleges
```
